## STEP3: Data cleaning & Exploratory Data Analysis (EDA)

### Remove the extra index column

### Detection and Removal of Missing Values

```python
null_table_2 = df_tweets_2.isnull()
null_table_2.tail()
```

```python
df_tweets_3 = df_tweets_2.dropna(how='any')
df_tweets_3.tail()
```

### Get a list of unique countries and their counts

### Remove Error rows

```python
df_tweets_4 = df_tweets_3[df_tweets_3.country != "Error"]
```

```python
df_tweets_4['country'].unique()
```

```python
df_tweets_4.describe()
```

### Remove Europe rows

```python
df_tweets_4 = df_tweets_4[df_tweets_4.country != "Europe"]
```

```python
df_tweets_4.describe()
```

### Standardize countries with multiple names 

```python
def standardize_countries(country):
  country = country.replace("United Kingdom", "UK")
  country = country.replace("United States","USA")
  country = country.replace ("Dubai - United Arab Emirates","UAE")
  country = country.replace ("- ŸÖÿµŸÅÿ≠M-37 - ÿ£ÿ®Ÿà ÿ∏ÿ®Ÿä - United Arab Emirates","UAE")
  country = country.replace("Jeddah Saudi Arabia", "Saudi Arabia")
  return country
```

```python
df_tweets_4["country"] = df_tweets_4["country"].apply(lambda x: standardize_countries(x))
```

```python
df_tweets_4['country'].unique()
```

### Breaking down the sentiment column

```python
df_tweets_4[['neg','neu','pos','compound']] = df_tweets_4.sentiment.str.split(",",expand=True,)
```

```python
df_tweets_4.head()
```

### Cleaning the new sentiment score columns

```python
def clean_sentiment_scores(score):
    score = re.sub ("\{*['][\w]*['']:*[\s]*", "", score)
    return score               
```

```python
df_tweets_4["pos"] = df_tweets_4["pos"].apply(lambda x: clean_sentiment_scores(x))
```

```python
df_tweets_4["neu"] = df_tweets_4["neu"].apply(lambda x: clean_sentiment_scores(x))
```

```python
df_tweets_4["neg"] = df_tweets_4["neg"].apply(lambda x: clean_sentiment_scores(x))
```

```python
df_tweets_4["compound"] = df_tweets_4["compound"].apply(lambda x: clean_sentiment_scores(x))
```

```python
def clean_sentiment_scores_2(score):
    score = re.sub ("\}", "", score)
    return score 
```

```python
df_tweets_4["compound"] = df_tweets_4["compound"].apply(lambda x: clean_sentiment_scores_2(x))
```

```python
df_tweets_4.head()
```

### Check and update data types
