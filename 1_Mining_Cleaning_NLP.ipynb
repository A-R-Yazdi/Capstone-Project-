### nltk.sentiment.vader
VADER ( Valence Aware Dictionary for Sentiment Reasoning) is a model used for text sentiment analysis that is sensitive to both polarity (positive/negative) and intensity (strength) of emotion. 

Since VADER is pretrained, you can get results more quickly than with many other analyzers. However, VADER is best suited for language used in social media, like short sentences with some slang and abbreviations. 

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

### lexicon:
VADER belongs to a type of sentiment analysis that is based on lexicons of sentiment-related words. In this approach, each of the words in the lexicon is rated as to whether it is positive or negative, and in many cases, how positive or negative.

## STEP 2: Pulling the data from Twitter

```python
accesstoken = config.get('twitter','accesstoken')
accesstokensecret = config.get('twitter','accesstokensecret')
apikey = config.get('twitter', 'apikey')
apisecretkey = config.get('twitter','apisecretkey')
```

```python
auth = tw.OAuthHandler(apikey,apisecretkey)
auth.set_access_token(accesstoken,accesstokensecret)
api = tw.API(auth, wait_on_rate_limit= True)
```

```python
search_words = ["coronavirus vaccine", "covid 19 vaccine", "covid vaccine"]
date_since = "2020-01-01"
```

```python
tweets = tw.Cursor(api.search, q = search_words, lang = 'en', since = date_since).items(1400)
```

```python
tweets
```

```python
tweet_details = [[tweet.geo, tweet.text, tweet.user.screen_name, tweet.user.location, tweet.created_at] for tweet in tweets]
```

```python
tweet_df = pd.DataFrame(data = tweet_details, columns= ('geo','text','user','location','created_at'))
```

## STEP 3: Data cleaning

```python

def clean_tweets(text):
  text = re.sub("RT @[\w]*:", " ", text)
  text = re.sub("@[\w]*","", text)
  text = re.sub("https?:/[A-Za-z0-9./]*","",text)
  text = re.sub("\n","",text)
  return text

```

```python
tweet_df["text"] = tweet_df["text"].apply(lambda x: clean_tweets(x))
```

## STEP 4: Natural Language Processing

### Named-entity recognition

```python
nlp = spacy.load('en_core_web_sm')
```

```python
tweet_df['text'].apply(lambda x: [print ("\tText : {}, Entity : {}".format(ent.text, ent.label_)) if (not ent.text.startswith("#")) else "" for ent in nlp(x).ents])
```

```python
tweet_df["entity"] = tweet_df['text'].apply(lambda x: [(ent.text, ent.label_) if (not ent.text.startswith("#")) else "" for ent in nlp(x).ents])
```

```python
tweet_df.head()
```

### Sentiment Analysis

```python
sid = SentimentIntensityAnalyzer()
```

```python
tweet_df['sentiment'] = tweet_df['text'].apply( lambda x: sid.polarity_scores(x))
```

```python
!pip install -U googlemaps
import googlemaps
```

```python
gmaps = googlemaps.Client(key = config.get('twitter','googleapikey'))
```

```python
geocode_results = gmaps.geocode(tweet_df['location'][4])
print(geocode_results)
print(geocode_results[0]['formatted_address'])
print(geocode_results[0]['formatted_address'].split(",")[-1].strip())
```

```python
def get_country (input):
  try:
    output = gmaps.geocode(input)[0]['formatted_address'].split(",")[-1].strip()
  except:
    output = "Error"
  return output
```

```python
tweet_df['country'] = tweet_df ['location'].apply (lambda x: "" if (not x.strip()) else get_country(x))
```
