## STEP3: Exploratory Data Analysis (EDA)

## Most Common Words

```python
data_dtm_2_T = data_dtm_2.transpose()
print(data_dtm_2_T.shape)
data_dtm_2_T.head()
```

### Find the top words tweeted in each country

```python
dict_all = {}
for c in data_dtm_2_T.columns:
    top = data_dtm_2_T[c].sort_values(ascending=False).head(10)
    dict_all[c]= list(zip(top.index, top.values))

dict_all
```

### Select top 5 countries with most tweets from the Document_Term Matrix (dtm)

```python
top5_list = ['USA','UK','Canada', 'Australia', 'New Zealand']
data_top5_dtm  = data_dtm_2[data_dtm_2.index.isin(top5_list)]
data_top5_dtm
```

#### Transpose the DataFrame

```python
data_top5_dtm_T = data_top5_dtm.transpose()
print(data_top5_dtm_T.shape)
data_top5_dtm_T.head()
```

### Find the top words tweeted in five eyes countries


```python
dict_five = {}
for c in data_top5_dtm_T.columns:
    top = data_top5_dtm_T[c].sort_values(ascending=False).head(20)
    dict_five[c]= list(zip(top.index, top.values))

dict_five
```

```python
dict_five_1 = {}
```

```python
for country, top_words in dict_five.items():
    print(country)
    count_list = []
    for word,count in top_words:
        count_list.append(count)
    print(count_list)
    dict_five_1[country]= count_list
    print('---')
```
